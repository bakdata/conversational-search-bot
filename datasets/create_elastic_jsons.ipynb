```python
import pandas as pd
import json
```

```python
books = pd.read_csv('booksummaries.txt', sep='\t')
books.columns = pd.Series(['wiki_id', 'freebase_id', 'title', 'author', 'publication_date', 'genres', 'summary'])
books = books.fillna('')
books
```

```python
json_array = []

for i, row in books.iterrows():

    year = str(row['publication_date']) if len(str(row['publication_date'])) == 4 else str(row['publication_date'])[:5]

    if len(row['genres']):
        genres = [v for k, v in json.loads(row['genres']).items()]
             
    doc = {
        'id': row['wiki_id'],
        'title': row['title'],
        'author': row['author'],
        'publication_year': year,
        'genres': genres,
        'summary': row['summary']
    }
    
    json_array.append(json.dumps({ "index" : { "_id" : row['wiki_id'] } }))
    json_array.append(json.dumps(doc))


f = open(f"booksummaries.json", "w+")
f.write("\n".join(json_array) + '\n')
```

```python
movies = pd.read_csv('movies/IMDb movies.csv')
movies = movies.fillna('')
```

```python
json_array = []

for i, row in movies.iterrows():
    if len(row['actors']):
        actors = row['actors'].split(",")

    doc = {
        'id': row['imdb_title_id'],
        'title': row['title'],
        'director': row['director'],
        'publication_year': row['year'],
        'genres': row['genre'],
        'summary': row['description'],
        'actors': actors
    }
    
    json_array.append(json.dumps({ "index" : { "_id" : row['imdb_title_id'] } }))
    json_array.append(json.dumps(doc))


f = open(f"movies.json", "w+")
f.write("\n".join(json_array) + '\n')
```

```python
ratings = pd.read_csv('movies/IMDb ratings.csv')
ratings
```

```python
json_array = []

for i, row in ratings.iterrows():
    doc = {
        'id': row['imdb_title_id'],
        'mean_rating': row['mean_vote'],
        'total_votes': row['total_votes']
    }
    
    json_array.append(json.dumps({ "index" : { "_id" : row['imdb_title_id'] } }))
    json_array.append(json.dumps(doc))


f = open(f"ratings.json", "w+")
f.write("\n".join(json_array) + '\n')
```
